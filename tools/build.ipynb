```python
import numpy as np
import pandas as pd
```

```python
# dataset grezzo
df = pd.read_csv(
    "dataset.csv",
    parse_dates=[
        "data", "pubblicazione", "popolazione_giorno",
        "diagnosi_start", "diagnosi_end",
        "ricoveri_start", "ricoveri_end",
        "intensive_start", "intensive_end",
        "decessi_start", "decessi_end"
    ],
    dtype={
        'popolazione': np.object,
        'diagnosi': np.object,
        'ricoveri': np.object,
        'intensive': np.object,
        'decessi': np.object,
    }
)
df
```

```python
# integer casting
df["popolazione"] = df.popolazione.str.replace(".", "", regex=False).astype(int)
df["diagnosi"] = df.diagnosi.str.replace(".", "", regex=False).astype(int)
df["ricoveri"] = df.ricoveri.str.replace(".", "", regex=False).astype(int)
df["intensive"] = df.intensive.str.replace(".", "", regex=False).astype(int)
df["decessi"] = df.decessi.str.replace(".", "", regex=False).astype(int)
```

```python
df.info()
```

```python
# popolazione totale per data e fascia
popolazione = df.groupby(["data", "fascia"])[["popolazione"]].sum().reset_index()
popolazione
```

```python
# aggiunta popolazione totale per fascia e data
for data in df.data.unique():
    for fascia in df.fascia.unique():
        _pop = popolazione.loc[(popolazione.data==data)&(popolazione.fascia==fascia), "popolazione"].values[0]
        df.loc[(df.data==data)&(df.fascia==fascia), "totale"] = _pop
df["totale"] = df.totale.astype(int)
```

```python
# calcolo percentuale di popolazione per data, fascia e stato vaccinale
df["percentuale_popolazione"] = df.popolazione / df.totale
```

```python
# ordinamento per data e fascia
df.sort_values(by=["data", "fascia"], inplace=True)
```

```python
# calcolo dei tassi per data, fascia e stato vaccinale
for y in ["diagnosi", "ricoveri", "intensive", "decessi"]:
    df[f"percentuale_{y}"] = df[y] / df.popolazione
```

```python
# salvataggio report completo
df.to_csv("../reports_iss_vaccini.csv", index=False)
```
